```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(12,8))
plt.title("Gas price over Time (in USD)", fontdict={'fontweight': 'bold', 'fontsize': 20})


for column in gas.columns[1:]:
    if column=='France':
        plt.plot(gas.Year,gas[column],'b.--',label=column)
    else:
        plt.plot(gas.Year,gas[column],label=column)
plt.xticks(gas.Year[::3].to_list()+[2011])

plt.xlabel('Years')
plt.ylabel('US $')
plt.legend()
plt.tight_layout()

plt.show()
```

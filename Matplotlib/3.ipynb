```python
import matplotlib.pyplot as plt

# Assuming 'fifa' is your DataFrame and 'Weight' is a column in it

light = fifa.loc[fifa.Weight < 125].count()[0]
light_med = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
med = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
medium_high = fifa.loc[fifa.Weight >= 200].count()[0]

weights_counts = [light, light_med, med, medium, medium_high]
weights_labels = ['light', 'light_med', 'med', 'medium', 'medium_high']

plt.pie(weights_counts, labels=weights_labels, autopct='%1.1f%%')
plt.title('Weight Distribution of FIFA Players')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'fifa' is your DataFrame and 'Weight' is a column in it
plt.style.use('ggplot')
# Convert the 'Weight' column to numeric, coercing errors to NaN
fifa['Weight'] = pd.to_numeric(fifa['Weight'], errors='coerce')

# Drop rows with NaN values in the 'Weight' column (if any)
fifa = fifa.dropna(subset=['Weight'])

# Calculate the counts of each weight category
light = fifa.loc[fifa.Weight < 125].count()[0]
light_med = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
med = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
medium_high = fifa.loc[fifa.Weight >= 200].count()[0]

# List of counts
weights_counts = [light, light_med, med, medium, medium_high]
# Corresponding labels
weights_labels = ['light', 'light_med', 'med', 'medium', 'medium_high']
explode=(.4,.2,0,0,0.5)
# Plot the pie chart with counts and labels
plt.pie(weights_counts, labels=weights_labels, autopct='.%2f %%',pctdistance=0.8,explode=explode)
plt.title('Weight Distribution of FIFA Players')
plt.tight_layout
plt.savefig('fifadata_pie.png')
plt.show()
```

```python
plt.style.use('default')
barcelona=fifa.loc[fifa.Club=="FC Barcelona"]['Overall']
plt.figure(figsize=(5,8))
madrid=fifa.loc[fifa.Club=="Real Madrid"]['Overall']
plt.title("Pro Soccer Team Comparsion",fontdict={'color':"red"})
plt.ylabel("FIFA Overall Rating")


labels=['FC Barcelona','Real Madrid']
boxes=plt.boxplot([barcelona,madrid],labels=labels,patch_artist=True,medianprops={'linewidth':2})
for box in boxes['boxes']:
    box.set(color='blue',linewidth=2)
    box.set(facecolor='#e0e0e0',linewidth=2)
plt.tight_layout()
plt.savefig('teams_boxplot.png')
plt.show()
```

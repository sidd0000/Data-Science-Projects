```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
plt.figure(figsize=(10,6),dpi=100)


# Plotting the data with customizations
# plt.plot(x, y, label='2x', color='y', linewidth=3, marker='o',markersize=14,linestyle='--',markeredgecolor='r')
plt.plot(x,y,'ro-',label='2x',markeredgecolor='black')
#line no 2
x2=np.arange(0,4.5,0.5)

plt.plot(x2[:6],x2[:6]**2,'b.-',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'b--')
# Setting the title with font properties
plt.title("First graph", fontdict={'fontsize': 40, 'fontname': 'Comic Sans MS', 'color': 'blue'})

# Setting the x and y labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Setting the x-ticks from 0 to 4
plt.xticks(np.arange(0, 9))

# Setting the y-ticks to even numbers
plt.yticks(np.arange(0, 10, 2))

# Displaying the legend
plt.legend()
plt.tight_layout()
# plt.tight_layout()

plt.savefig('final.png')


# Displaying the plot
plt.show()
```

##bar chart

```python
import matplotlib.pyplot as plt
import numpy as np

l = ['A', 'B', 'C']
v = [1, 4, 2]
patterns = ['/', '*', 'O']

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the bar plot
bars = plt.bar(l, v, color='r', hatch='.')

# Set y-ticks
plt.yticks(np.arange(0, 6))

# Apply hatch patterns to each bar
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

# Save the figure
plt.savefig('bar.png')

# Show the plot
plt.show()
```

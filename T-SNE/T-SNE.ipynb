```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
d0=pd.read_csv('../PCA/mnist_train.csv')
l=d0['label']
d=d0.drop("label",axis=1)


data=d.head(15000)
labels=l.head(15000)

from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
from sklearn.manifold import TSNE
print(standardized_data)
```

```python
from sklearn.manifold import TSNE

# Picking the top 1000 points as TSNE takes a lot of time for 15K points
data_1000 = standardized_data[0:42000,:]
labels_1000 = labels[0:42000]
model=TSNE(n_components=2,random_state=0,perplexity=10,max_iter=5000)
tsne_data=model.fit_transform(data_1000)
tsne_data=np.vstack((tsne_data.T,labels_1000)).T   
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))
sns.FacetGrid(tsne_df, hue="label", height=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()


```

```python
from sklearn.manifold import TSNE

# Picking the top 1000 points as TSNE takes a lot of time for 15K points
data_1000 = standardized_data[0:1000,:]
labels_1000 = labels[0:1000]
model=TSNE(n_components=2,random_state=0,perplexity=20,max_iter=5000)
tsne_data=model.fit_transform(data_1000)
tsne_data=np.vstack((tsne_data.T,labels_1000)).T   
tsne_df=pd.DataFrame(data=tsne_data,columns=("Dim_1","Dim_2","label"))
sns.FacetGrid(tsne_df, hue="label", height=6).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

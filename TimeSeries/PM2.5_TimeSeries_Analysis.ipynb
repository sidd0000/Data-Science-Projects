```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2
```

```python
a = np.random.multivariate_normal([0, 6], [[2, 1], [1, 1.5]], size=200)
b = np.random.multivariate_normal([2, 0], [[1, -1], [-1, 3]], size=400)
c = np.random.multivariate_normal([6, 4], [[5, 0], [0, 1.2]], size=250)

# Concatenating the arrays
z = np.concatenate((a, b, c))

# Shuffling the combined array
np.random.shuffle(z)

# Plotting the data
plt.scatter(z[:, 0], z[:, 1])
plt.show()
k=3#no of clusters
```

```python
centroids,labels=kmeans2(z,k,minit="points")
print(f"Centrooids:{centroids}")
print("z:",z)


```

```python
plt.style.use('ggplot')

cluster_1=z[labels==0]
cluster_2=z[labels==1]
cluster_3=z[labels==2]
plt.scatter(cluster_1[:,0],cluster_1[:,1],label='cluster_1')
plt.scatter(cluster_2[:,0],cluster_2[:,1],label='cluster_2')
plt.scatter(cluster_3[:,0],cluster_3[:,1],label='cluster_3',marker='o')
plt.scatter(centroids[:,0],centroids[:,1],label='centroids',color='black')
plt.legend()
plt.show()


```

```python
x=np.arange(11)
y=np.array([2.0,1.9,1.7,1.5,0.5,0.0,0.8,2.0,0.9,0.2,2.0])
plt.plot(x,y,"bo:")
plt.show()  
```

```python
from scipy.interpolate import interp1d
predict=interp1d(x,y,kind='quadratic')
x2=np.linspace(0,10,100)

y2=np.array([predict(x) for x in x2])
plt.plot(x, y, 'bo:', label='Original data')
plt.plot(x2, y2, 'ro:', label='Interpolated data')
plt.legend()
plt.show()

# Get user input and make prediction
predicted_value = float(input("Enter the number for which you want to see the prediction: "))
print(f"x: {predicted_value}, y: {predict(predicted_value)}")
```

```python
import numpy as np
import scipy.linalg

# Example eigenvectors and eigenvalues (replace these with your actual values)
eg_vec = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])
eg_val = [2, 3, 4]

# Compute the inverse of the eigenvector matrix
P_inverse = scipy.linalg.inv(eg_vec)

# Construct the diagonal matrix D from eigenvalues
D = np.zeros((3, 3))
for index, eg_value in enumerate(eg_val):
    D[index, index] = eg_value

# Calculate the matrix product P @ D @ P_inverse
result = eg_vec @ D @ P_inverse

print("Resulting Matrix:")
print(result)
```

```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
    a = x[0]
    b = x[1]
    return a**2 + b**2

def constraint1(x):
    a = x[0]
    b = x[1]
    return a + b - 100

def constraint2(x):
    a = x[0]
    b = x[1]
    return a * b - 100

bounds = [(10, 50), (10, 50)]

cons = [
    {
        'type': "eq",
        'fun': constraint1
    },
    {   
        'type': "ineq",
        'fun': constraint2
    }
]

x0 = np.array([20, 50])
solution = minimize(objective, x0, method='SLSQP', constraints=cons, bounds=bounds)

print(solution)
print("Optimized values of x:", solution.x)
```

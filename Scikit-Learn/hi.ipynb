```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris

# Load a sample dataset
data = load_iris()
X, y = data.data, data.target

# Define the model
model = RandomForestClassifier()

# Define the K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate the model
scores = cross_val_score(model, X, y, cv=kf)

print(f'Cross-Validation Scores: {scores}')
print(f'Mean Score: {scores.mean()}')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics

# Larger Sample Data
emails = [
    "free money now", "win a prize", "exclusive offer just for you",
    "meeting agenda for tomorrow", "project update due next week",
    "company news and updates", "free ticket", "buy now",
    "discount on your favorite products", "important project deadline",
    "team meeting at 3 PM", "workshop schedule"
]

labels = ["spam", "spam", "spam", "not spam", "not spam",
          "not spam", "spam", "spam", "spam", "not spam",
          "not spam", "not spam"]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(emails, labels, test_size=0.2, random_state=42)

# Print split data for debugging
print("Training Data:")
for email, label in zip(X_train, y_train):
    print(f"{email} -> {label}")

print("\nTest Data:")
for email, label in zip(X_test, y_test):
    print(f"{email} -> {label}")

# Create a pipeline that combines a CountVectorizer with a Naive Bayes classifier
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Evaluate the model
accuracy = metrics.accuracy_score(y_test, predictions)
print(f"\nAccuracy: {accuracy:.2f}")

# Example predictions
new_emails = ["win money now", "update on project", "buy tickets"]
predicted_labels = model.predict(new_emails)
print("\nPredictions for new emails:")
for email, label in zip(new_emails, predicted_labels):
    print(f"'{email}' -> {label}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
emails = [
    "free money now", "win a prize", "exclusive offer just for you",
    "meeting agenda for tomorrow", "project update due next week",
    "company news and updates", "free ticket", "buy now",
    "discount on your favorite products", "important project deadline",
    "team meeting at 3 PM", "workshop schedule"
]
labels = ["spam", "spam", "spam", "not spam", "not spam",
          "not spam", "spam", "spam", "spam", "not spam",
          "not spam", "not spam"]
X_train,X_test,y_train,y_test=train_test_split(emails,labels,test_size=0.2,random_state=42)
model = make_pipeline(TfidfVectorizer(ngram_range=(1, 2)), MultinomialNB())
model.fit(X_train, y_train)
predicitons=model.predict(X_test)
accuracy=metrics.accuracy_score(y_test,predicitons)
print(f"Accuracy: {accuracy:.2f}")

# Example predictions
new_emails = ["win money now", "update on project", "buy tickets","workshop schedule"]
predicted_labels = model.predict(new_emails)
print("\nPredictions for new emails:")
for email, label in zip(new_emails, predicted_labels):
    print(f"'{email}' -> {label}")
```

```python
# Sample dataset
import numpy as np
import matplotlib.pyplot as plt

# Heights (in inches) and Weights (in pounds)
X = np.array([58, 59, 60, 61, 62, 63, 64, 65, 66, 67]).reshape(-1, 1)  # Heights (feature)
y = np.array([115, 117, 120, 123, 126, 129, 132, 135, 139, 142])  # Weights (target)

```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Heights (independent variable) and Weights (dependent variable)
X = np.array([58, 59, 60, 61, 62, 63, 64, 65, 66, 67]).reshape(-1, 1)  # Heights (in inches)
y = np.array([115, 117, 120, 123, 126, 129, 132, 135, 139, 142])        # Weights (in pounds)

# Create a Linear Regression model
model = LinearRegression()

# Train the model using the dataset (fit the line)
model.fit(X, y)

# Get the parameters of the model (slope and intercept)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (β1): {slope}")
print(f"Intercept (β0): {intercept}")

# Make predictions (for the given heights)
y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Predict weight for a person with height of 68 inches
height_new = np.array([[68]])
predicted_weight = model.predict(height_new)
print(f"Predicted weight for 68 inches: {predicted_weight[0]:.2f} pounds")

# Plot the original data and the regression line
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.title('Height vs Weight - Linear Regression')
plt.legend()
plt.show()
```
